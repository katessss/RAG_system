```python
import sqlite3
import json
from src.databases import get_sqlite_conn, get_chroma_collection
conn =get_sqlite_conn()
conn.row_factory = sqlite3.Row # Чтобы обращаться к полям по именам
cursor = conn.cursor()

cursor.execute("SELECT original_content, metadata FROM docs_fts")
rows = cursor.fetchall()

chunks = []
for row in rows:
    meta = json.loads(row["metadata"])
    chunks.append({
        "content": row["original_content"],
        "metadata": meta,
        "type": meta.get("type", "text")
    })

conn.close()
print(f"Из SQLite загружено: {len(chunks)} чанков")
# chunks
```

```python
from pathlib import Path

src_root = Path(".")
out_dir = Path("py_as_txt")
out_dir.mkdir(parents=True, exist_ok=True)

for py_path in src_root.rglob("*.py"):
    # игнорируем служебные папки
    if any(part in {"venv", ".venv", ".git", "__pycache__"} for part in py_path.parts):
        continue

    # создаём уникальное имя файла
    flat_name = "__".join(py_path.relative_to(src_root).parts)
    flat_name = flat_name.replace(".py", ".txt")

    txt_path = out_dir / flat_name

    txt_path.write_text(
        py_path.read_text(encoding="utf-8", errors="replace"),
        encoding="utf-8"
    )

print("Готово ✅ Всё в одной папке")
```

```python
from pathlib import Path
import shutil

folder = Path("py_as_txt")  # имя папки

if folder.exists():
    for item in folder.iterdir():
        if item.is_file() or item.is_symlink():
            item.unlink()
        elif item.is_dir():
            shutil.rmtree(item)

print("Папка очищена")
```
